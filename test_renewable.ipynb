The graph above shows a simple economic diagram (cost vs. demand). The blue slope shows that the more quantity of electricity is needed, the more the price of marginal cost increases, because indeed, the last unit of production costs more to produce. 

The demand curve represents the marginal benefits consumers see for using one more unit of electricity. It is decreasing with growing Q because, imagine if we go from one demand Q1 to a lower demand Q2, the cost for the customer will increase (electricity more rare for the customer). 

If we consider pollution, we have social costs for the producer. The blue line goes up to the dashed blue line. The new equilibrium is found in Q'. The resulting market price is found at P', reflecting the added marginal costs of pollution in the market outcome. 

We see here in this example that if we had added the pollution (externality) in consideration, the marginal cost for producer increase. If carbon is taxed, it could help renewable energy to become more viable economically.
